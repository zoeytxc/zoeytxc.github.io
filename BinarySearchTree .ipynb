```python
'''
A binary search tree can be implemented using three arrays:
one for the left and right pointers, and one for data

The data 17, 8, 4, 12, 22, 19, 14, 5, 30, 25 can be stored
as follows:

        left   data   right
bst[0]  1      17     4
bst[1]  2      8      3
bst[2]  -1     4      7
bst[3]  -1     12     6
bst[4]  5      22     8
bst[5]  -1     19     -1
bst[6]  -1     14     -1
bst[7]  -1     5      -1
bst[8]  9      30     -1
bst[9]  -1     25     -1  

Implement the binary search tree using the above representation.
You should use appropriate classes with the following methods

initialise()
insert()
search()
delete()
update()
inorder() / display()
preorder()
postorder()
maximum()
minimum()

'''
class BST:
    Max=11
    
    def __init__(self):
        self.data=[]
        self.left=[]
        self.right=[]
        for i in range(self.Max):
            self.data.append(-1)
            self.left.append(-1)
            self.right.append(-1)
        self.top=-1

    def insert(self,new):
        self.top+=1
        self.data[self.top]=new
        i=0
 
        if self.top!=0:
            while not((new<self.data[i] and self.left[i]==-1) or (new>self.data[i] and self.right[i]==-1)):
                if(new<self.data[i]):i=self.left[i]
                elif(new>self.data[i]):i=self.right[i]
     
            if(new<self.data[i]) and self.top!=0: self.left[i]=self.top
            if(new>self.data[i]) and self.top!=0: self.right[i]=self.top
            


    def search(self,target):
        i=0
        while (self.data[i]!=target and i<self.top):
            i+=1
        if self.data[i]==target: return "Found"
        else: return "Not found"

    def lookup(self,target): #return current and node parent
        i=0
        while (self.data[i]!=target and i<self.top):
            i+=1
        if (self.data[i]!=target): i=-1
        j=0
        while (self.left[j]!=i and self.right[j]!=i and j<i):
            j+=1
        
        if (self.left[j]!=i and self.right[j]!=i): j=-1
        return i,j

    def delete(self,target):
        # 1:delete leave node(just delete)
        # 2:delete node with one child(point parent to node child)
        # 3: delete node with 2 children(replace with inorder predecessor or inorder successor)
        i,j=self.lookup(target)
        if i!=-1:
            # case 1:
            if (self.left[i]==-1) and (self.right[i]==-1):
                if self.left[j]==i:
                    self.left[j]=-1
                else:
                    self.right[j]=-1
            # case2:
            elif (self.left[i]==-1)!=(self.right[i]==-1):
                if self.left[i]!=-1:
                    n=self.left[i]
                else:
                    n=self.right[i]
                if self.left[j]==i:
                    self.left[j]=n
                else:
                    self.right[j]=n
            # case 3:
            else:
                j=i
                k=self.right[i]
                while self.left[k]!=-1:
                    j=k
                    k=self.left[k]
                #replace node data with successor data
                self.data[i]=self.data[k]
                #fix successor's parent node child
                if self.left[j]==k:
                    self.left[j]=self.right[k]
                else:
                    self.right[j]=self.right[k]

    def  preorder(self,i=0): #ååºéå
        print(self.data[i], end=' ')
        if self.left[i]!=-1:
            self.preorder(self.left[i])
        if self.right[i]!=-1:
            self.preorder(self.right[i])
        
 
        

    def  postorder(self,i=0): #ååºéå
        if self.left[i]!=-1:
            self.postorder(self.left[i])
        if self.right[i]!=-1:
            self.postorder(self.right[i])
        print(self.data[i], end=' ')

    def display(self,i=0): #inorder traversal
        if self.left[i]!=-1:
            self.display(self.left[i])
        print(self.data[i], end=' ')
        if self.right[i]!=-1:
            self.display(self.right[i])
        
    def reverse(self,i=0):
        if self.right[i]!=-1:
            self.reverse(self.right[i])
        print(self.data[i], end=' ')
        if self.left[i]!=-1:
            self.reverse(self.left[i])
        

    def minimum(self):
        mini=32767
        for i in range(self.top+1):
            if mini>self.data[i]: mini=self.data[i]
        print("Minimum is ", mini)

    def maximum(self):
        maxi=-1
        for i in range(self.top+1):
            if maxi<self.data[i]: maxi=self.data[i]
        print("Maximum is ", maxi)
        
tree=BST()

tree.insert(17)
tree.insert(8)
tree.insert(4)
tree.insert(12)
tree.insert(22)
tree.insert(19)
tree.insert(14)
tree.insert(5)
tree.insert(30)
tree.insert(25)
tree.delete(22)
print(tree.left)
print(tree.data)
print(tree.right)
```
