```python
'''
The Fibonacci sequence was originally used as a basic model for rabbit population growth.
The Fibonacci sequence goes like this:

0,1,1,2,3,5,8,13,...

The nth term is the sum of the previous two terms.
In the above example, the next term would be 21 (8+13).

Create a Fibonacci sequence generator up to the 50th term. 
The program will then prompt the user for n (up to 50).
E.g
Enter n: 7
Fibonacci number is 8
'''
a=0
b=1
i=1
n=int(input("Enter n: "))
while i<n:
    c=a+b
    a=b
    b=c
    i+=1
print("Fibonacci number is", a)
```

```python
'''
Write a program that will store names into an array.
As a new name is entered it will be added to the end of the array.
The user can keep adding names until they enter the sentinel value 'ok'
Once this has been done the program will display any duplicate names.
E.g.
Enter name: Bill
Enter name: Mary
Enter name: Anisha
Enter name: Mary
Enter name: Bill
Enter name: Mary
Enter name: exit
Duplicates: Bill Mary 

'''
d={}
r=[]
u=input("Enter name: ")
while u!="ok":
    if u in d.keys():
        if u not in r:
            r.append(u)
    else:
        d[u]="0"
    u=input("Enter name: ")
print("Duplicates: ",end="")
for i in r:
    print(i,end=" ")
```

```python
''' Create a simple treasure hunt game. Create a two-dimensional 10 by 10 array of integers. In a random position in the array store the number 1. Repeat Get the user to enter coordinates where he/she think the treasure is. If there is a 1 at this position display 'Bingo!'. Until they find the treasure ''' 
import random
a=[[0 for x in range(10)] for y in range(10)] 
for i in range(10): 
    for j in range(10): 
        a[i][j]=random.randint(2,15) 
x=random.randint(0,9) 
y=random.randint(0,9) 
a[x][y]=1
while True: 
    ux=int(input("Location X: ")) 
    uy=int(input("Location Y: ")) 
    if a[ux][uy]==1: 
        print("Bingo") 
        break 
    else: print("Try Again")

```

```python
'''
Write a program that will generate a random playing card 
e.g. '9 Hearts', 'Queen Spades' 
when the return key is pressed.

Rather than generate a random number from 1 to 52. 
Create two random numbers â one for the suit and one for the card.

However we do not want the same card drawn twice.
Update this program by using an array to prevent the same card being 
dealt twice from the pack of cards.
Convert this code into a procedure 'DealCard' that will display the
card dealt or 'No more cards'.
Call your procedure 53 times (use a loop!).
'''
from random import randint
a=[[0 for x in range(4)] for y in range(14)] 
def cards(i):
    if i==52:
        print("No more cards")
       
    else:
        f=False
        while f==False:
            map={11:'Jack',12:"Queen",13:"King"}
            card={1:"Spades",2:"Hearts",3:"Clubs",0:"Diamonds"}
            r=randint(1,13)
            c=randint(0,3)
            if a[r][c]==0:
                a[r][c]==1
                f=True
        if r<=10:print(r,end=' ')
        else: print(map[r],end=' ')
        print(card[c])

for i in range(53):
    cards(i)
    
```

```python
'''
Play a game that draws two random cards (see above).
The player then decides to draw or stick.
If the score goes over 21 the player loses.
Keep drawing until the player sticks.
After the player sticks draw two computer cards.
If the player beats the score they win.

Extension: Aces can be 1 or 11!
The number used is whichever gets the highest score. 
'''
from random import randint
a=[[0 for x in range(4)] for y in range(14)] 
def cards():
    f=False
    while f==False:
        map={11:'Jack',12:"Queen",13:"King"}
        card={1:"Spades",2:"Hearts",3:"Clubs",0:"Diamonds"}
        r=randint(1,13)
        c=randint(0,3)
        if a[r][c]==0:
            a[r][c]==1
            f=True
    if r<=10:print(r,end=' ')
    else: print(map[r],end=' ')
    print(card[c])
    
    return r,c

r1,c1=cards()
r2,c2=cards()
rr=r1+r2
u=input("Stick or draw?")
while u!="stick":
    r,c=cards()
    rr+=r
    if rr>21:
        print("You Lose!!!")
        break
    u=input("Stick or draw?")
if u=='stick':
    ur1,uc1=cards()
    ur2,uc2=cards()
    ur=ur1+ur2
    if ur<rr:
        print("You Win!!!")
    else:
        print("You Lose!!!")
```

```python
'''
The computer generates a 4-digit code
The user types in a 4-digit code. His/her guess.
The computer tells them how many digits they guessed correctly
'''
from random import randint
a=[]
for i in range(4):
    a.append(randint(0,9))
u=input("Enter the code: ")

r=0
for i in range(4):
    if int(u[i]) in a: r+=1
if r<2:
    print(r,"digit correct")
else:
    print(r,"digits correct")
```
